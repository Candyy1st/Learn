```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
x = sns.pairplot(tips, hue='sex')
x.map_diag(sns.displot)
x.map_upper(sns.kdeplot)
```

```python
sns.countplot(x='smoker', data=tips)
```

```python
sns.pointplot(x='smoker', y='tip', data=tips)
```

```python
sns.catplot(x='smoker', y='tip', data=tips, kind='box')
```

```python
sns.swarmplot(x='smoker', y='tip', data=tips, hue='sex', palette='cool')
```

```python
sns.stripplot(x='smoker', y='tip', data=tips, hue='sex', palette='cool')
```

```python
sns.barplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sns.boxplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex')
```
