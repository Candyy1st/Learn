```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tip = tips.corr()
sns.heatmap(tip, annot=True, lw=1, linecolor='blue')
```

```python
tip = tips.pivot_table(index='day', columns='tip', values='total_bill')
tip
```

```python
flight = flights.pivot_table(index='month', columns='year', values='passengers')
flight
```

```python
sns.heatmap(flight, lw=1)
```

```python
sns.countplot(x='smoker', data=tips)
```
