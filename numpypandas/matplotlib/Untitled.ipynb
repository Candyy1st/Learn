```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,20)
y = x * x
z = x + y
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Testing Matplotlib')
```

```python
plt.subplot(3,2,1)
plt.plot(x,x)
plt.subplot(3,2,2)
plt.plot(x,y)
plt.subplot(3,2,5)
plt.plot(x,z)
plt.subplot(3,2,6)
plt.plot(z,z)
plt.tight_layout()
```
