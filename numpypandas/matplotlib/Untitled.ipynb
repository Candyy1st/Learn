```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,20)
y = x * x
z = x + y
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Testing Matplotlib')
```

```python
plt.subplot(3,2,1)
plt.plot(x,y)
plt.title('Plot')

plt.subplot(3,2,2)
plt.scatter(x,y)
plt.title('Dot')

plt.subplot(3,2,3)
plt.hist(x,y)
plt.title('Histogram')

plt.subplot(3,2,4)
plt.bar(x,y)
plt.title('Bar Graph')

plt.subplot(3,2,5)
plt.fill(x,y)
plt.title('Fill')

plt.tight_layout()
```

```python
plt.polar(x,y)
```

```python
fig = plt.figure()
x1 = fig.add_axes([0.1,0.1,0.8,0.8])
x1.plot(x,y,y,z,y,x)
x2 = fig.add_axes([0.2,0.2,0.3,0.3])
x2.plot(y,x)
x3 = fig.add_axes([0.5,0.5,0.3,0.3])
x3.plot(x,y)

x1.set_xlabel('x-axis')
x1.set_ylabel('y-axis')
x1.set_title('Graph')
x2.set_xlabel('x-axis')
x2.set_ylabel('y-axis')
x2.set_title('Graph')

x2.plot(x,y)
```

```python
fig, axes = plt.subplots(2,3, figsize=(10,4)) #add figsize to change them 

axes[0,0].plot(x,y)
axes[0,1].plot(y,x)
axes[1,1].plot(x,y)
axes[1,2].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
x1 = fig.add_axes([0,0,1,1])
x1.set_xlabel('x-axis')
x1.set_ylabel('y-axis')
x1.set_title('Graph')

#add color to edit color
#add linestyle to change style line
#add linewidth/lw to change width line
#add alpha for transparancy
x1.plot(x,y, color='b', linestyle='dashed', lw=3)

#use marker to add marker
#use markerfacecolor for change size marker
#use markerfacecolor for change color marker
#use markeredgecolor for change color line marker
x1.plot(y,x, color='b', marker='o', markersize=10, markerfacecolor='r', markeredgecolor='g')
```
