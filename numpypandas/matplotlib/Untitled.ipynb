```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,20)
y = x * x
z = x + y
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Testing Matplotlib')
```

```python
plt.subplot(3,2,1)
plt.plot(x,x)
plt.title('Plot')
plt.subplot(3,2,2)
plt.scatter(x,y)
plt.title('Dot')
plt.subplot(3,2,3)
plt.hist(x,z)
plt.title('Histogram')
plt.subplot(3,2,4)
plt.bar(z,z)
plt.title('Bar Graph')
plt.subplot(3,2,5)
plt.fill(z,z)
plt.title('Fill')
plt.subplot(3,2,6)
plt.polar(z,z)
plt.title('polar')
plt.tight_layout()
```
